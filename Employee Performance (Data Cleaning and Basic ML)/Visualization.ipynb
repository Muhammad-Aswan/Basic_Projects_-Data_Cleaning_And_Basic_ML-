```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = pd.read_csv("Cleaned_EMP.csv")
df.drop(columns='Unnamed: 0', inplace=True)
df.head()
```

```python
plt.figure()
sns.histplot(df["Salary"], bins=5)
plt.title("Salary Distribution")
plt.show()
```

```python
plt.figure()
sns.barplot(x="Department", y="Salary", data=df)
plt.title("Average Salary by Department")
plt.show()
```

```python
plt.figure()
sns.scatterplot(x="PerformanceScore", y="Salary", data=df)
plt.title("Performance vs Salary")
plt.show()
```

```python
plt.figure()
sns.histplot(df["Bonus"], bins=5)
plt.title("Bonus Distribution")
plt.show()
```

```python
plt.figure()
sns.countplot(x="AttritionRisk", data=df)
plt.title("Attrition Risk Count")
plt.show()
```

```python
plt.figure()
sns.boxplot(x="Seniority", y="Salary", data=df)
plt.title("Salary by Seniority Level")
plt.show()
```

```python
plt.figure()

corr = df.select_dtypes(include="number").corr()
sns.heatmap(corr, annot=True)

plt.title("Correlation Matrix")
plt.show()
```
