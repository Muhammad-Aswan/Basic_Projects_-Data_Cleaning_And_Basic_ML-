```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Messy_Dataset.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df["Quantity"] = df["Quantity"].fillna(df["Quantity"].median())
df["UnitPrice"] = df["UnitPrice"].fillna(df["UnitPrice"].median())
df["City"] = df["City"].fillna("Unknown")
df["ReturnFlag"] = df["ReturnFlag"].fillna(0)
df
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df["OrderDate"] = pd.to_datetime(df["OrderDate"])
```

# Feature Engineering

```python
df['TotalRevenue'] = df['Quantity'] * df['UnitPrice']
df.head()
```

```python
df['HighOrder'] = np.where(df["TotalRevenue"] > 50000,1,0)
df.head()
```

```python
df["City"].value_counts()
```

```python
df["CityGroup"] = df["City"].map({
    "Karachi":"Tier1",
    "Lahore":"Tier2",
    "Islamabad":"Tier2",
    "Unknown":"Tier3"
})
df
```

# EDA (Exploratory Data Analysis)
