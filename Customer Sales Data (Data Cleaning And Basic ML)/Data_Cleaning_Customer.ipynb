```python
df["Age"] = df["Age"].fillna(df["Age"].median())
df["Country"] = df["Country"].fillna("Unknown")
df["PurchaseAmount"] = df["PurchaseAmount"].fillna(df["PurchaseAmount"].median())
df["Discount"] = df["Discount"].fillna(0)
df["Rating"] = df["Rating"].fillna(df["Rating"].mean())
df
```

```python
df.isnull().sum()
```

```python
df["PurchaseDate"] = pd.to_datetime(df["PurchaseDate"])
df
```

# Feature Engineering

```python
df["NetPurchase"] = df["PurchaseAmount"] - (df["PurchaseAmount"]*df["Discount"] / 100)
df
```

```python
df["AgeGroup"] = np.where(df["Age"] < 30, "Young", "Adult")
df
```

```python
df["CustomerScore"] = (df["NetPurchase"] / 10000) + df["Rating"]
df
```
