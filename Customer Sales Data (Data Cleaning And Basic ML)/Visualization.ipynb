```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv("Cleaned_Dataset.csv")
df.head()
df.drop(columns="Unnamed: 0", inplace=True)
df
```

```python
print('============================== SHAPE (ROWS, COLUMNS) ====================================== \n\n')
print(df.shape)
print('============================== DESCRIBING  ====================================== n\n')
print(df.describe(), '\n\n')
print('============================== INFORMATION ======================================\n\n\n')
print(df.info())
```

#  VISUALIZATION

```python
sns.barplot(x="Country",y="NetPurchase", data=df)
plt.title("Sales by Country")
plt.show()
```

```python
plt.figure()

sns.histplot(df["NetPurchase"], bins=20, kde=True)

plt.title("Net Purchase Distribution")
plt.xlabel("Net Purchase Amount")
plt.ylabel("Frequency")

plt.show()
```

```python
plt.figure()
corr = df.select_dtypes(include=["number"]).corr()
sns.heatmap(corr, annot=True)
plt.title("Correlation Matrix")
plt.show()
```

```python
sns.boxplot(x="Product", y="NetPurchase", data=df)
plt.title("Profit by Category")
plt.show()
```

```python
buyers = df.groupby("Product")["CustomerID"].nunique()

plt.figure()
sns.barplot(x=buyers.index, y=buyers.values)
plt.title("Number of Buyers by Product")
plt.ylabel("Buyers")
plt.show()
```

```python
revenue = df.groupby("Product")["NetPurchase"].sum()

plt.figure()
sns.barplot(x=revenue.index, y=revenue.values)
plt.title("Net Purchase by Product")
plt.ylabel("Net Purchase")
plt.show()
```

# Insight

Most of the money is coming from India. Customers are of two types — some buy medium price things, and some buy expensive things. There is almost no one buying in between.If someone buys more, their net purchase is obviously higher.
Customers who give good ratings usually buy more and have higher customer score.
Age doesn’t matter much here. Selling Laptops makes the most money per sale. Phones make the least profit.The number of customers is almost equal for all three products. No product is more popular than others in terms of buyer count.Even though same number of people bought each product, Laptops brought the most total money. Phones brought the least total money.

“Laptops are making the most profit and total money. India is our best market. Phones are not performing well. Customers who give good ratings usually spend more money.”
