```python
df['Age'].fillna(df["Age"].median(), inplace=True)
df['Salary'].fillna(df['Salary'].median(), inplace=True)
df['City'].fillna("Unknown", inplace=True)
df['Bonus'].fillna(0, inplace=True)
df
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
df = df.drop_duplicates()
```

```python
df['JoinDate'] = pd.to_datetime(df["JoinDate"])
df
```

# Feature Engineering 

```python
df["TenureYears"] = 2026 - df["JoinDate"].dt.year
df
```

```python
df["TCompension"] = df["Salary"] + df["Bonus"]
df
```

```python
def perf(score):
    if score >= 9:
        return "Excellent"
    elif score >= 7.5:
        return "Good"
    else:
        return "Average"
    
df['PStatus'] = df['PerformanceScore'].apply(perf)
```

```python
df
```

```python
def level(exp):
    if exp >= 15:
        return "Senior"
    elif exp >= 7:
        return "Mid"
    else:
        return "Junior"

df["Seniority"] = df["Experience"].apply(level)
df
```
