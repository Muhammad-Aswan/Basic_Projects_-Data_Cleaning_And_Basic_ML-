```python
df["TransactionAmount"] = df["TransactionAmount"].astype(float)
df["AccountBalance"] = df["AccountBalance"].astype(float)
```

# Feature Engineering

```python
df["HighValue"] = np.where(df["TransactionAmount"] > 100000, 1,0)
df
```

```python
df["ForeignRisk"] = df["IsForeignTransaction"]
```

```python
df
```

```python
df["NightRisk"] = np.where(df["TimeOfDay"] == "Night", 1, 0)
```
