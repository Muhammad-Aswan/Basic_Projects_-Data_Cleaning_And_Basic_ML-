```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Messy_dataset.csv")
df.head()
```

```python
df["JoinDate"] = pd.to_datetime(df["JoinDate"])
df["LastPurchaseDate"] = pd.to_datetime(df["LastPurchaseDate"])
```

# Feature Engineering

```python
df["CustomerLifetime"] = (df["LastPurchaseDate"] - df["JoinDate"]).dt.days
df.head()
```

```python
today = pd.to_datetime("2023-07-20")
df["Recency"] = (today - df["LastPurchaseDate"]).dt.days
df.head()
```

```python
df["PurchaseFrequency"] = df["TotalOrders"] / df["CustomerLifetime"]
df
```

```python
df["MonetaryValue"] = df["TotalSpent"] / df["TotalOrders"]
df.head()
```

```python
df["CLV"] = df["PurchaseFrequency"] * df["MonetaryValue"] * df["CustomerLifetime"]
df.head()
```

```python
df["CustomerSegment"] = np.where(df["CLV"] > 200000, "VIP",
                         np.where(df["CLV"] > 100000, "Loyal", "Regular"))
df.head()
```
